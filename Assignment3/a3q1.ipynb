```python
import numpy as np
```

```python
def graient_descent(x,y):
    mcurr=bcurr=0
    n=len(x)
    iterations = 1000
    learningrate=0.001
    for i in range(0,1000):
        print(mcurr,bcurr,i)
        y_predicted= x*mcurr+bcurr                   #y=2X+4
        md= -(2/n)*sum(x*(y-y_predicted))
        bd= -(2/n)*sum(y-y_predicted)
        mcurr=mcurr - md*learningrate
        bcurr=bcurr - bd*learningrate
        
```

```python
x=np.array([1,3,6])
y=np.array([6,10,16])
```

```python
graient_descent(x,y)
```

# Covariance and Varience between XY

```python
xm=sum(x)/len(x)
ym=sum(y)/len(y)
cov=(sum((x-xm)*(y-ym)))/len(x)

var=(sum((x-xm)*(x-xm)))/len(x)
```
