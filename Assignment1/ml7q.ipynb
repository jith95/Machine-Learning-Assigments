```python
import pandas as pd
dataset1 = pd.read_csv('iris.csv')
print(dataset1.head(5))
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import time
time1 = []
f1 = .1
train_time = []
test_time = []
data_size = []
for f in range (0,10):
    d1 = dataset1.sample(frac = f1)
    data_size.append(f1)
    d1 = dataset1.drop("species",axis=1)
    t1 = time.time()
    x_train, x_test, y_train, y_test = train_test_split(d1,dataset1['species'] , test_size=0.3)
    c = tree.DecisionTreeClassifier()
    c = c.fit(x_train,y_train)
    t2 = time.time()
    train_time.append(t2-t1)
    
    t1 = time.time()
    prediction = c.predict(x_test)
    a = accuracy_score(y_test, prediction, normalize=True, sample_weight=None)
    t2 = time.time()
    test_time.append(t2-t1)
    f1 = f1 + .1
```

```python
import matplotlib.pyplot as plt
plt.plot(data_size, train_time) 
plt.xlabel('Data_size in %') 
plt.ylabel('Training_time') 
plt.title("The variation of training time as a function of data size")
plt.show()
```

```python
plt.plot(data_size, test_time) 
plt.xlabel('data_size in %') 
plt.ylabel('test_time') 
plt.title("variation of testing time with varying data size")
plt.show()
```
